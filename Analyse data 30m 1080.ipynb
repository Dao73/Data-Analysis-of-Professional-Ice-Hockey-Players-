```python
import pandas as pd

csv_file_path = "C:\Python\Dossier csv\Receuil donn√©es 1080 - Test 30m v1.csv"

# Charger le fichier CSV dans un DataFrame
df = pd.read_csv(csv_file_path)

print(df)
```

```python
# Grouper les donn√©es par session et sprint, puis calculer les statistiques descriptives pour chaque groupe
stats_descriptives = df.groupby(['Session'])[['Temps 30 m', 'Temps 5 m', 'F0', 'V0', 'Pmax']].describe()

# Appliquer le style avec une alternance de gris et de blanc
styled_table = stats_descriptives.style.set_properties(**{'background-color': 'white',
                                                         'color': 'black',
                                                         'border-color': 'black'})

# Appliquer une couleur de fond diff√©rente pour les lignes impaires
styled_table = styled_table.apply(lambda x: ['background-color: #f2f2f2' if i % 2 == 0 else '' for i, _ in enumerate(x)], axis=1)

# Ajouter un titre au tableau avec une police plus grande
styled_table.set_caption('<h2 style="font-size: 20px">Tableau des statistiques descriptives par session et sprint</h2>')

# Afficher le tableau de statistiques descriptives stylis√© avec un titre
styled_table
```

```python
# Grouper les donn√©es par session et sprint, puis calculer les statistiques descriptives pour chaque groupe
stats_descriptives = df.groupby(['Session'])[['Temps 30 m']].describe()

# Appliquer le style avec une alternance de gris et de blanc
styled_table = stats_descriptives.style.set_properties(**{'background-color': 'white',
                                                         'color': 'black',
                                                         'border-color': 'black'})

# Appliquer une couleur de fond diff√©rente pour les lignes impaires
styled_table = styled_table.apply(lambda x: ['background-color: #f2f2f2' if i % 2 == 0 else '' for i, _ in enumerate(x)], axis=1)

# Ajouter un titre au tableau avec une police plus grande
styled_table.set_caption('<h2 style="font-size: 20px">Tableau des statistiques descriptives par session et sprint</h2>')

# Afficher le tableau de statistiques descriptives stylis√© avec un titre
styled_table
```

```python
from tabulate import tabulate

classement_temps_5m = df.sort_values(by='Temps 5 m')

# Ajouter une colonne pour les rangs
classement_temps_5m['Rang'] = range(1, len(classement_temps_5m) + 1)

# Formatage du tableau avec des couleurs altern√©es
tableau = tabulate(classement_temps_5m[['Rang', 'Nom', 'Temps 5 m']], headers='keys', tablefmt='fancy_grid', showindex=False, numalign="center", stralign="center")

# Afficher le tableau
print(tableau)
```

```python
from tabulate import tabulate

classement_temps_5m = df.sort_values(by='Temps 30 m')

# Ajouter une colonne pour les rangs
classement_temps_5m['Rang'] = range(1, len(classement_temps_5m) + 1)

# Formatage du tableau avec des couleurs altern√©es
tableau = tabulate(classement_temps_5m[['Rang', 'Nom', 'Temps 30 m']], headers='keys', tablefmt='fancy_grid', showindex=False, numalign="center", stralign="center")

# Afficher le tableau
print(tableau)
```

```python
# Trier le DataFrame par la colonne "Temps 5 m" de mani√®re descendante (du plus grand au plus petit)
classement_temps_5m = df.sort_values(by='Pmax', ascending=False)

# Ajouter une colonne pour les rangs
classement_temps_5m['Rang'] = range(1, len(classement_temps_5m) + 1)

# Formatage du tableau avec des couleurs altern√©es
tableau = tabulate(classement_temps_5m[['Rang', 'Nom', 'Pmax']], headers='keys', tablefmt='fancy_grid', showindex=False, numalign="center", stralign="center")

# Afficher le tableau
print(tableau)
```
