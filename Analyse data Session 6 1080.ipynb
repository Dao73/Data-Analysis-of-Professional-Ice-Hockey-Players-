```python
import pandas as pd

csv_file_path = "C:\Python\Dossier csv\Receuil données 1080 - Session 6.csv"
# Charger le fichier CSV dans un DataFrame
df = pd.read_csv(csv_file_path)

print(df.to_string())
```

```python
import pandas as pd

# Charger votre DataFrame depuis Google Sheets ou tout autre source
# Assumons que votre DataFrame est nommé df et la colonne est 'Temps'

# Convertir la colonne 'Temps' en données numériques
df['Temps'] = pd.to_numeric(df['Temps'], errors='coerce')

# Afficher le DataFrame pour vérification
print(df.to_string())

```

```python
import matplotlib.pyplot as plt

# Calculer la moyenne de chaque sprint pour chaque session
sprint_means = df.groupby(['Session', 'Série','7 m'])['Puissance'].mean().unstack()

# Créer le graphique
plt.figure(figsize=(10, 6))
for session in sprint_means.index:
    plt.plot(sprint_means.columns, sprint_means.loc[session], marker='o', label=f'Session {session}')

# Ajouter des étiquettes et un titre
plt.xlabel('Série')
plt.ylabel('Puissance moyenne')
plt.title('Moyenne des puissances par sprint pour toutes les sessions')
plt.legend()
plt.grid(True)
plt.show()
```
